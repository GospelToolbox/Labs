```python
# Need 13
songs = [
  "Goodness Of God",
  "How Rich A Treasure We Possess",
  "Sovereign",
  "O Great God",
  "Sovereign Over Us",
  "Hail The Day That Christ Arose",
  "O Lord My Rock And My Redeemer",
  "God You're So Good",
  "He Will Hold Me Fast",
  "Yet Not I But Through Christ In Me",
  "The Greatness Of Our God",
  "His Mercy Is More",
  "Worthy Of More",
  "Hosanna (Praise Is Rising)",
  "King Of Kings",
  "Jesus Messiah",
]
```

```python
from numpy.random import choice
import pprint
import random

song_count = len(songs)
print("{} songs".format(song_count))

probs_abs = [4] * song_count
set_count = 9

def get_rel_probs(probs_abs):
    return list(map(lambda x: x/sum(probs_abs), probs_abs))

def pick_song(probs_rel):
    return choice(songs, 1, p=probs_rel)[0]

def pick_set(probs_abs, length=4):
    set_probs = list(probs_abs)
    picked = []
    
    while len(picked) < length:
        rel_probs = get_rel_probs(set_probs)
        song = pick_song(rel_probs)
        idx = songs.index(song)
        picked.append(song)
        set_probs[idx] = 0
        
    return picked

def get_next_set(probs_abs, length=4):
    next_set = pick_set(probs_abs, length)
    for song in next_set:
        idx = songs.index(song)
        probs_abs[idx] -= 1
        
    return (next_set, probs_abs)

sets = []

current_probs = list(probs_abs)

time_since_last_5 = 0

while len(sets) < set_count:
    is_fiver = 0 #random.random() > 0.85 - (.15 * time_since_last_5)
    
    time_since_last_5 += 1
    if is_fiver:
        time_since_last_5 = 0
        
    song_count = 5 if is_fiver else 4
    (next_set, current_probs) = get_next_set(current_probs, song_count)
    sets.append(next_set)
    
    
for n in range(set_count):
    print("- # Set {}:".format(n+1))
    
    for song in sets[n]:
        print("  - {}".format(song))
    
    print()
```
